### Reading text file 

```python
f = open("bio.txt", "r")
print(f.read())
```

### Writing text file
- "a" - Append - will append to the end of the file
- "w" - Write - will overwrite any existing content

```python
f = open("bio.txt", "a")
f.write("Working remotely from india.")
f.close()
```

```python
f = open("bio.txt", "r")
print(f.read())
```

```python
f = open("bio.txt", "w")
f.write("In wolf scraping team")
f.close()
```

```python
f = open("bio.txt", "r")
print(f.read())
```

### Create a New Text File
- "x" - Create - will create a file, returns an error if the file exist
- "a" - Append - will create a file if the specified file does not exist
- "w" - Write - will create a file if the specified file does not exist

```python
f = open("bio.txt", "x")
```

```python
f = open("demo.txt", "x")
```

```python
f = open("demo.txt", "w")
f.write("This a demo file created for testing purpose")
f.close()
```

```python
f = open("demo.txt", "r")
print(f.read())
```

### Creating json file

```python
import json
```

```python
x = {
  "name": "John",
  "age": 30,
  "city": "New York"
}

# convert into JSON:
y = json.dumps(x)

# the result is a JSON string:
print(y)
```

```python
f = open("sample.json", "x")
```

### Writing json file

```python
f = open("sample.json", "w")
f.write(y)
f.close()
```

### Reading json file

```python
f = open("sample.json", "r")
print(f.read())
```
